```python
model = ['GSVM']
instance = list(range(120,150,10))
Q_add_GSVM=[70]
Q_init_GSVM=[30]


for i in range (len(model)):
    for k in range (len(Q_init_GSVM)):
        Q_init=Q_init_GSVM[k]
        Q_max=Q_init_GSVM[k]+Q_add_GSVM[k]
        for j in range (len(instance)):
            configdict_mlca = OrderedDict([('SATS_domain_name',model[i]),
                                ('SATS_auction_instance_seed',instance[j]+2),
                                ('bidders',0),
                                ('items',0),
                                ('bidder_ids',0),
                                ('Qinit',Q_init),
                                ('Qmax',Q_max),
                                ('Qround',0),
                                ('epochs' , 300),
                                ('batch_size' , 30),
                                ('lra',0.02),
                                ('regularization',1e-5),
                                ('regularization_type' , 'l1_l2'),
                                ('layer_N' , [16,16]),
                                ('layer_R' ,[16,16] ),
                                ('layer_L' , [16,16]),
                                ('NN_parameters',[]),
                                ('bigM',6000),
                                ('warm_start',False),
                                ('MIP_parameters',[]),
                                ('scaler',None),
                                ('init_bids_and_fitted_scaler',[None,None]),
                                ('return_allocation',True),
                                ('return_payments',True),
                                ('calc_efficiency_per_iteration',True)])
            print(configdict_mlca['SATS_auction_instance_seed'])
            set_value_model_parameters(configdict_mlca)    
            res = mlca_mechanism(configdict = configdict_mlca)


            filename =configdict_mlca['SATS_domain_name']+'_'+str(configdict_mlca['Qinit'])+'_'+str(configdict_mlca['Qmax'])+'_'+str(configdict_mlca['SATS_auction_instance_seed'])+'_timelimit_20_'+'mip_torch'+'.pkl'
            dir_file = dirname+'/'+filename
            _file=OrderedDict([('configdict_mlca',configdict_mlca),('result',res)])
            save_result(dir_file,_file)

```
